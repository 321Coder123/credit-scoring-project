```python
import os

# Chemins des fichiers
path_csv = '../data/raw/application_train.csv'
path_parquet = '../data/processed/application_train.parquet'

# Récupérer la taille en octets et convertir en Mo (Mégaoctets)
size_csv = os.path.getsize(path_csv) / (1024 * 1024)
size_parquet = os.path.getsize(path_parquet) / (1024 * 1024)

print(f"Taille CSV : {size_csv:.2f} Mo")
print(f"Taille Parquet : {size_parquet:.2f} Mo")
print(f"Réduction de stockage : {100 - (size_parquet / size_csv * 100):.2f} %")
```
